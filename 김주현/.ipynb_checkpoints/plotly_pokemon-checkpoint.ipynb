```python
import pandas as pd
import numpy as np
import os
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
df = pd.read_csv("./pokemon.csv")
```

```python
# df.isnull().sum()
```

```python
# nan value preprocess
df.loc[df['type2'].isnull(), 'type2'] = df['type1']
df = df.fillna(0)
```

```python
stats = ['attack', 'defense', 'hp', 'sp_attack', 'sp_defense', 'speed']
against = ['against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water']
```

```python
df.loc[df["is_legendary"]==1, "is_legendary"] = "Y"
df.loc[df["is_legendary"]==0, "is_legendary"] = "N"
df['avg'] = df[stats].mean(axis=1)
df['agst_avg'] = df[against].mean(axis=1)
df["stddev"]=df[stats].std(axis=1)
df["tmp"] = "_"
df["type_total"] = df["type1"]+df["tmp"]+df["type2"]
# df['generation'] = df['generation'].astype(str)+df["tmp"]
df = df.drop("tmp", axis=1)
```

```python
DF_legend = df.loc[df["is_legendary"]=="Y"]
DF_nonlegend = df.loc[df["is_legendary"]=="N"]
```

```python
def compare2pokemon(x,y):
    x = df[df["name"] == x]
    y = df[df["name"] == y]

    trace0 = go.Scatterpolar(
      r = [x['attack'].values[0],x['defense'].values[0],x['hp'].values[0],x['sp_attack'].values[0]
           ,x['sp_defense'].values[0],x['speed'].values[0],x['attack'].values[0]],
      theta = ['attack', 'defense', 'hp', 'sp_attack', 'sp_defense', 'speed','attack'],
      fill = 'toself',
      name = x.name.values[0]
    )

    trace1 = go.Scatterpolar(
      r = [y['attack'].values[0],y['defense'].values[0],y['hp'].values[0],y['sp_attack'].values[0]
           ,y['sp_defense'].values[0],y['speed'].values[0],y['attack'].values[0]],
      theta = ['attack', 'defense', 'hp', 'sp_attack', 'sp_defense', 'speed','attack'],
      fill = 'toself',
      name = y.name.values[0]
    )

    data = [trace0, trace1]

    layout = go.Layout(
      polar = dict(
        radialaxis = dict(
          visible = True,
          range = [0, 200]
        )
      ),
      showlegend = True,
      title = "{}-{} vs {}-{}".format(x.name.values[0],x.type_total.values[0],y.name.values[0],y.type_total.values[0])
    )
    fig = go.Figure(data=data, layout=layout)
    fig.show()
#     iplot(fig, filename = "Two Pokemon stats")
```

```python
compare2pokemon("Rayquaza","Mewtwo")
```

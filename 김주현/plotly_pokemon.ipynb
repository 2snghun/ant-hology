```python
import pandas as pd
import numpy as np
import os
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("./pokemon.csv")

df.loc[df["is_legendary"]==1, "is_legendary"] = "Y"
df.loc[df["is_legendary"]==0, "is_legendary"] = "N"
df["tmp"] = "_"
df["type_total"] = df["type1"]+df["tmp"]+df["type2"]
df['generation'] = df['generation'].astype(str)+df["tmp"]
df = df.drop("tmp", axis=1)
```

```python
df.head(3)
```

```python
df.columns
```

```python
fig = px.violin(df
             , x="is_legendary"
             , y="base_total"
             , color="generation"
             , hover_data=df.columns
             #, points="all"
            )
fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Violin(x=df['type1'][ df['is_legendary'] == "Y" ],
                        y=df['base_total'][ df['is_legendary'] == "Y" ],
                        legendgroup='Yes', scalegroup='Yes', name='Yes',
                        side='negative',
                        line_color='blue')
             )
fig.add_trace(go.Violin(x=df['type1'][ df['is_legendary'] == "N" ],
                        y=df['base_total'][ df['is_legendary'] == "N"  ],
                        legendgroup='No', scalegroup='No', name='No',
                        side='positive',
                        line_color='orange')
             )
fig.update_traces(meanline_visible=True)
fig.update_layout(violingap=0, violinmode='overlay')
fig.show()
```

```python
fig = px.scatter_polar(df, r="base_total", theta="generation",
                          color="is_legendary", symbol="is_legendary"
                      )
fig.show()
```

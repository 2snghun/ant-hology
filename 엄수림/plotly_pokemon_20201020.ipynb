- Min-Max Normalized 뒤 Radar Chart 그리기

```python
normalized_df = df[['pokedex_number' ,'name', 'is_legendary', 'base_total', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]

for col in ['base_total', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']:
    normalized_df['n'+col] = (normalized_df[col] - normalized_df[col].min()) / (normalized_df[col].max() - normalized_df[col].min())

normalized_df.head()
```

```python
import plotly.graph_objects as go

categories = ['nbase_total', 'nattack', 'ndefense', 'nsp_attack', 'nsp_defense', 'nspeed']
data = normalized_df[(normalized_df.base_total == 700) & (normalized_df.is_legendary == 0)][['pokedex_number' ,'name'] + categories].reset_index(drop=True)
data['nspeed'] = 1 - data['nspeed']

fig = go.Figure()

for i in range(len(data)):
    fig.add_trace(go.Scatterpolar(
          r=data[categories].iloc[i].tolist(),
          theta=categories,
          fill='toself',
          name=data['name'].iloc[i]
    ))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
  showlegend=False
)

fig.show()
```
